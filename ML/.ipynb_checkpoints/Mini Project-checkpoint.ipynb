```python
import numpy as np 
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns 

# Read data files
train_data = pd.read_csv('E:/Coding/ML/train.csv')
test_data = pd.read_csv('E:/Coding/ML/test.csv')

# Check for missing values
print(train_data.isnull().sum())
print(test_data.isnull().sum())

# Drop 'PassengerId' as it's not useful
train_data.drop('PassengerId', axis=1, inplace=True)
test_data.drop('PassengerId', axis=1, inplace=True)

# Drop 'Name' as it's not useful
train_data.drop('Name', axis=1, inplace=True)
test_data.drop('Name', axis=1, inplace=True)

# Plot pie chart for 'Sex'
train_data['Sex'].value_counts().plot(
    kind="pie",
    explode=[0.05 for x in train_data['Sex'].dropna().unique()],
    autopct='%1.1f%%',
    shadow=True
)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()

# Convert 'Sex' to numerical
train_data['Sex'] = train_data['Sex'].map({'male': 1, 'female': 0})
test_data['Sex'] = test_data['Sex'].map({'male': 1, 'female': 0})

# Handle missing 'Age' by filling with mean
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())

# Drop 'Cabin' due to many missing values
train_data.drop('Cabin', axis=1, inplace=True)
test_data.drop('Cabin', axis=1, inplace=True)

# Drop rows with missing 'Embarked'
train_data.dropna(subset=['Embarked'], inplace=True)

# One-Hot Encode 'Embarked'
train_data = pd.get_dummies(train_data, columns=['Embarked'], prefix='Embarked')
test_data = pd.get_dummies(test_data, columns=['Embarked'], prefix='Embarked')

# Handle 'Ticket' column (Option 1: Dropping it)
train_data.drop('Ticket', axis=1, inplace=True)
test_data.drop('Ticket', axis=1, inplace=True)

# Alternatively, handle 'Ticket' as shown in the solutions above

# Verify no missing values
print(train_data.isnull().sum())
print(test_data.isnull().sum())

# Prepare features and target variable
X = train_data.drop('Survived', axis=1)
y = train_data['Survived']  # Changed to Series instead of DataFrame

# Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)

# Initialize and train the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(max_depth=4, random_state=42)
rf_clf.fit(X_train, y_train)

# Predictions and Evaluation
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix

print('Max depth:', 4)
prediction = rf_clf.predict(X_test)
print('Model accuracy: {0:0.4f}'.format(accuracy_score(y_test, prediction)))
print('Precision score: {0:0.4f}'.format(precision_score(y_test, prediction)))
print('Confusion Matrix:\n', confusion_matrix(y_test, prediction))

# Feature Importances
feature_names = X.columns.to_list()
importances = rf_clf.feature_importances_

print("\nFeature Importances:")
for feature, importance in zip(feature_names, importances):
    print(f'{feature}: {round(importance, 4) * 100}%')
```
